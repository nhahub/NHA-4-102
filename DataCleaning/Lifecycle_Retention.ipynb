```python
import pandas as pd
import numpy as np
```

```python
Lifecycle_Retention=pd.read_excel(r"../hire_to_retire_analytics.xlsx",sheet_name="Lifecycle_Retention")
Lifecycle_Retention.head()
```

```python
Lifecycle_Retention_new=Lifecycle_Retention.copy()
```

```python
Lifecycle_Retention_new.info()
```

```python
Lifecycle_Retention_new.describe()
```

```python
Lifecycle_Retention_new.shape
```

```python
Lifecycle_Retention_new.isnull().sum()
```

```python
Lifecycle_Retention_new["Exit_Reason"]=Lifecycle_Retention_new['Exit_Reason'].fillna("Still Active")
```

```python
Lifecycle_Retention_new["Termination_Date"]=Lifecycle_Retention_new['Termination_Date'].fillna(0)
```

```python
Lifecycle_Retention_new["Last_Promotion_Date"]=Lifecycle_Retention_new['Last_Promotion_Date'].fillna(0)
```

```python
Lifecycle_Retention_new["Avg_KPI_Career"]=Lifecycle_Retention_new['Avg_KPI_Career'].fillna(Lifecycle_Retention_new['Avg_KPI_Career'].median())
```

```python
Lifecycle_Retention_new.isnull().sum()
```

```python
Lifecycle_Retention_new.duplicated().sum()
```

```python
for col in Lifecycle_Retention_new.columns:
    if Lifecycle_Retention_new[col].dtype=="object":
        Lifecycle_Retention_new[col]=Lifecycle_Retention_new[col].astype("string")
```

```python
Lifecycle_Retention_new["Termination_Date"]=pd.to_datetime(Lifecycle_Retention_new["Termination_Date"],errors="coerce")
```

```python
Lifecycle_Retention_new["Last_Promotion_Date"]=pd.to_datetime(Lifecycle_Retention_new["Last_Promotion_Date"],errors="coerce")
```

```python
Lifecycle_Retention_new.info()
```

```python
Lifecycle_Retention_new.to_csv("../CleanedCSVs/Lifecycle_Retention_new.csv",index=False)
```
